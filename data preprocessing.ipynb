```python
for i in Indicators:
  v=np.mean(df3.loc[(df3['CountryName']=='Denmark') & (df3['IndicatorName']==i)]['Value'])
  final.loc[final['Countries']=='Denmark',i]=v
final.head()
```

```python
for j in Countries:
  for i in Indicators:
      v=np.mean(df3.loc[(df3['CountryName']==j) & (df3['IndicatorName']==i)]['Value'])
      final.loc[final['Countries']==j,i]=v
final.head()
```
